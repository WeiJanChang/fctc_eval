```python
import pandas as pd
from typing import Optional, List, Dict, Tuple
import pandas as pd
import numpy as np
from pathlib import Path
import collections 
```

```python
df = pd.read_excel('/Users/wei/UCD-MPH/MPH-Lecture:Modules/MPH Dissertation/Data/19_ratified_country.xlsx')
```

```python
def ratified_19_countries_layout_modified(df: pd.DataFrame,
                            column_drop: Optional[Path] = None,
                            save_path: Optional[Path] = None) -> pd.DataFrame:
    if column_drop is not None:
        df = df.drop(columns=column_drop)
    dy: dict = collections.defaultdict(list)
    group = df.groupby(['Country Name', 'Year', 'Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths',
                        'Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths',
                        'Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate',
                        'Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate',
                        'Ratified Year'])
    info: List[Tuple] = list(group.groups.keys())

    for i, it in enumerate(info):
        dy['Country Name'].append(it[0])
        dy['Year'].append(it[1])
        dy['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths'].append(it[2])
        dy['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths'].append(it[3])
        dy['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate'].append(it[4])
        dy['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate'].append(it[5])
        dy['Ratified Year'].append(it[6])


    _df = pd.DataFrame.from_dict(dy)
    country_values = ['Austria', 'Czechia','Costa Rica', 'Ecuador','Estonia', 'Georgia',
                      'Guatemala', 'Germany','Iceland','Kazakhstan', 'Lithuania',
                      'Latvia','Mauritius', 'Mexico',
                      'Netherlands','Singapore', 'Spain', 'Serbia', 'Slovenia']
    changed_df = _df.assign(
        AUT_CVD = _df.query("Country Name == 'Austria'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        AUT_TBC=_df.query("Country Name == 'Austria'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],
        
        CZE_CVD=_df.query("Country Name == 'Czechia'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        CZE_TBC=_df.query("Country Name == 'Czechia'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],
        
        CRI_CVD=_df.query("Country Name == 'Costa Rica'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        CRI_TBC=_df.query("Country Name == 'Costa Rica'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],
        
        ECU_CVD=_df.query("Country Name == 'Ecuador'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        ECU_TBC=_df.query("Country Name == 'Ecuador'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],
        
        EST_CVD=_df.query("Country Name == 'Estonia'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        EST_TBC=_df.query("Country Name == 'Estonia'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],
        
        GEO_CVD=_df.query("Country Name == 'Georgia'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        GEO_TBC=_df.query("Country Name == 'Georgia'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],
               
        GTM_CVD=_df.query("Country Name == 'Guatemala'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        GTM_TBC=_df.query("Country Name == 'Guatemala'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'],       

        DEU_CVD=_df.query("Country Name == 'Germany'")
        ['Female_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']
        ['Male_Total_Percentage_of_Cause_Specific_Deaths_Out_Of_Total_Deaths']['Ratified Year'],
        DEU_TBC=_df.query("Country Name == 'Germany'")
        ['Female_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']
        ['Male_Estimate_of_Current_Tobacco_Use_Prevalence_age_standardized_rate']['Ratified Year'])      
        
        
        
    changed_df.reset_index(drop=True, inplace=True)

    changed_df = changed_df.drop(['Country Name'], axis=1)
    changed_df = changed_df.groupby(['Year']).first().reset_index()
    print(changed_df)
    if save_path is not None:
        changed_df.to_excel(save_path)
    return changed_df
```
